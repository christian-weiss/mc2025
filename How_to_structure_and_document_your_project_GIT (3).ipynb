**update the web repository with the changes you have done in your local computer**
- Windows:
  - change directory into my_data_project (cd + path to my_data_project)
  - git add * (update the web repository with the changes you have done in your local computer)

- Mac:
  - cd ~/Desktop/my_data_project
  - git add *



image.png

Check the status of your changes
- Changes to be commited means you have added files but this are not commited to the remote repository yet.

**Now let's commit the changes**
Because this is your first time you will need to set your user name an mail before commiting

- Windows:
  - then you write and run: git commit -m "first upload"
  - then you write and run: git push

- Mac:
  - git commit -m "Your commit message here"
  - git push origin main
